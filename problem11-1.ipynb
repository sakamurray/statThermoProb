# 章末問題 11.1の解答例


温度$T=3000$ Kの酸素分子の比エンタルピーを計算しなさい．ただし，酸素分子の分子振動の振動数を$\nu = 4.73 \times 10^{13}$ 1/s，モル質量を$M=32.0$ kg/kmolとし，電子励起エネルギーを無視すること．また，ボルツマン定数は$k = 1.38 \times 10^{-23}$ J/K，プランク定数は$h = 6.63 \times 10^{-34}$ J·sとする．

酸素分子の比内部エネルギー$e$は，以下のように各エネルギーモードの寄与の和として表される．
$$e = e_\mathrm{trans} + e_\mathrm{rot} + e_\mathrm{vib}$$

ここで，$e_\mathrm{trans}$は並進エネルギーの寄与，$e_\mathrm{rot}$は回転エネルギーの寄与，$e_\mathrm{vib}$は振動エネルギーの寄与である．

まず，並進エネルギーと回転エネルギーの寄与はそれぞれ以下のように与えられる．

$$e_\mathrm{trans} = \frac{3}{2} R T$$
$$e_\mathrm{rot} = R T$$
ここで，$R$は酸素分子の気体定数であり，$R = k N_A/ M$で与えられる．ただし，$N_A$はアボガドロ数$6.022 \times 10^{26}$ 1/kmolである．


また，振動エネルギーの寄与は
$$e_\mathrm{vib} = R \left( \frac{h \nu}{k} \right) \frac{1}{\exp\left( \frac{h \nu}{k T} \right) - 1}$$

以上より，求める酸素分子の比エンタルピー$h$は次式によって与えられる．
$$h = e + RT$$


```python
import numpy as np

k = 1.38e-23 # J/k
h = 6.63e-34 # J·s
nu = 4.73e13 # 1/s
M = 32.0 # kg/kmol
T = 3000 # K
theta_v = h * nu / k # K, 振動特性温度　
NA = 6.022e26 # 1/kmol, アボガドロ数


R = k * NA / M # J/(kg·K), 気体定数

etrans = 3/2 * R * T
erot = R * T
evib = R * theta_v / (np.exp(theta_v / T) - 1)

e = etrans + erot + evib

h = e + R * T

print(f"比エンタルピー: {h/1e6:.2f} MJ/kg")
```
